```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress
plt.style.use('bmh')
```

```python
x, y = [], []
```

```python
for c in range(100):
    x.append(c)
    y.append(np.random.randint(9,999))
    pass

x = np.array(x)
y = np.array(y)
```

```python
p1 = np.polyfit(x, y, 1)
poly_1 = np.polyval(p1, x)

p2 = np.polyfit(x, y, 12)
poly_2 = np.polyval(p2, x)

p3 = np.polyfit(x, y, np.floor(len(x) * 0.15))
poly_3 = np.polyval(p3, x)

p1, p2, p3
```

```python
plt.figure(figsize=(18,6))
plt.plot(x, y)
plt.plot(x, poly_1)

plt.tight_layout()
```

```python
plt.figure(figsize=(18,6))
plt.plot(x, y)
plt.plot(x, poly_2)

plt.tight_layout()
```

```python
plt.figure(figsize=(18,6))
plt.plot(x, y)
plt.plot(x, poly_3)

plt.tight_layout()
```

```python
slope, intercept, rvalue, pvalue, stderr = linregress(x,y)

print(f'{slope}, {intercept}, {rvalue}, {pvalue}, {stderr}')
```
