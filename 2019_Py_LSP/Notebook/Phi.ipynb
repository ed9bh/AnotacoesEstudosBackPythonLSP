```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
Phi_list = []
i, j, c = 1, 1, 50
for _ in range(c):
    Phi_list.append([i, j])
    i, j = j, i + j
    pass
Phi_list
```

```python
plt.plot(Phi_list)
plt.show()
```

```python
last = 0
rest = False
plst = [0]
n = 0
for x in Phi_list:
    x1 = np.float64(x[1])
    x2 = np.float64(x[0])
    calc = np.true_divide(x1, x2)
    n += 1
    rest = calc == last
    last = calc
    print(f'{n:03} - {calc:0.72f} : {rest}')
    plst.append(calc)
    phi = calc
```

```python
plt.plot(plst)
plt.axhline(1, color='r')
plt.show()
```

```python
plt.plot([0, phi ** (3), phi], label='Phi ao Cubo')
plt.legend()
plt.show()
```
