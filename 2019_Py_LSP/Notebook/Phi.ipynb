```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
Phi_list = []
i, j, c = 1, 1, 50
for _ in range(c):
    Phi_list.append([i, j])
    i, j = j, i + j
    pass
Phi_list
```

```python
plt.plot(Phi_list)
plt.show()
```

```python
last = 0
rest = False
plst = []
plst_anti = []
n = 0
for x in Phi_list:
    x1 = np.float64(x[1])
    x2 = np.float64(x[0])
    calc = np.true_divide(x1, x2)
    n += 1
    rest = calc == last
    last = calc
    print(f'{n:03} - {calc:0.72f} : {rest}')
    plst.append(calc)
    plst_anti.append(calc * -1)
    phi = calc

plst = np.array(plst)
plst_anti = np.array(plst_anti)
```

```python
x = 1
x_mirror = []
for i in range(len(plst)):
    x -= 1
    x_mirror.append(x)

x_mirror = np.array(x_mirror)
x = x_mirror * -1
```

```python
plt.figure(figsize=(18,9))

plt.plot(x, plst - 1)
plt.plot(x, plst_anti + 1)

plt.plot(x_mirror, plst - 1)
plt.plot(x_mirror, plst_anti + 1)

plt.axhline(0, color='black', linewidth=0.9, linestyle=':')
plt.axvline(0, color='black', linewidth=0.9, linestyle=':')
plt.show()
```

```python
plt.figure(figsize=(18,9))

plt.plot(x - 24.5, plst - 1.5)
plt.plot(x - 24.5, plst_anti + 1.5)

plt.plot(x_mirror + 24.5, plst - 1.5)
plt.plot(x_mirror + 24.5, plst_anti + 1.5)

plt.show()
```

```python
plt.plot([0, phi ** (3), phi], label='Phi ao Cubo')
plt.legend()
plt.show()
```
