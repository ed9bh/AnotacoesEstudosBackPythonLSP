```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('bmh')
```

### Curvas de Bombas

```python
df_Curvas_de_Bombas_1 = pd.DataFrame()
```

```python
df_Curvas_de_Bombas_1['Qv'] = [0,10,20,30,40,50,60]
RPM_List = [1000, 700, 1100]
df_Curvas_de_Bombas_1['Hman_RPM' + str(RPM_List[0])] = [100,95,90,80,65,45,20]
df_Curvas_de_Bombas_1['Hman_RPM' + str(RPM_List[1])] = (df_Curvas_de_Bombas_1['Hman_RPM' + str(RPM_List[0])] * (RPM_List[1] ** 2)) / (RPM_List[0] ** 2)
df_Curvas_de_Bombas_1['Hman_RPM' + str(RPM_List[2])] = (df_Curvas_de_Bombas_1['Hman_RPM' + str(RPM_List[1])] * (RPM_List[2] ** 2)) / (RPM_List[1] ** 2)

df_Curvas_de_Bombas_1
```

```python
plt.plot(df_Curvas_de_Bombas_1['Qv'], df_Curvas_de_Bombas_1['Hman_RPM1000'], label='1000 RPM')
plt.plot(df_Curvas_de_Bombas_1['Qv'], df_Curvas_de_Bombas_1['Hman_RPM700'], label='700 RPM')
plt.plot(df_Curvas_de_Bombas_1['Qv'], df_Curvas_de_Bombas_1['Hman_RPM1100'], label='1100 RPM')

plt.xlabel('Qv - Vaz√£o Volum√©trica')
plt.ylabel('Hman - Altura Manom√©trica')

plt.title('Curvas de Bombas - Rota√ß√£o')

plt.legend()

plt.show()
```

```python
df_Curvas_de_Bombas_2 = pd.DataFrame()
```

```python
df_Curvas_de_Bombas_2['Qv'] = [0,10,20,30,40,50,60]
df_Curvas_de_Bombas_2['Hman_D=200mm'] = [100,95,90,80,65,45,20]
D200, D240, D150 = 200, 240, 150
df_Curvas_de_Bombas_2['Hman_D=240mm'] = (df_Curvas_de_Bombas_2['Hman_D=200mm'] * (D240 ** 2)) / (D200 ** 2)
df_Curvas_de_Bombas_2['Hman_D=150mm'] = (df_Curvas_de_Bombas_2['Hman_D=240mm'] * (D150 ** 2)) / (D240 ** 2)

df_Curvas_de_Bombas_2
```

```python
plt.plot(df_Curvas_de_Bombas_2['Qv'], df_Curvas_de_Bombas_2['Hman_D=200mm'], label='D=200')
plt.plot(df_Curvas_de_Bombas_2['Qv'], df_Curvas_de_Bombas_2['Hman_D=240mm'], label='D=240')
plt.plot(df_Curvas_de_Bombas_2['Qv'], df_Curvas_de_Bombas_2['Hman_D=150mm'], label='D=150')

plt.xlabel('Qv - Vaz√£o Volum√©trica')
plt.ylabel('Hman - Altura Manom√©trica')

plt.title('Curvas de Bombas - Rotor')

plt.legend()

plt.show()
```
