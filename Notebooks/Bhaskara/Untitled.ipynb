```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
baskara = pd.DataFrame({
    'a':[-1,2,-3,6,-2],
    'b':[1,3,-2,-5,5],
    'c':[11,-11,2,-3,7]
})

baskara
```

```python
baskara['Delta'] = (baskara['b'] ** 2) - (4 * baskara['a'] * baskara['c'])

baskara['X1'] = ((baskara['b'] * -1) + (baskara['Delta'] ** 0.5)) / (2 * baskara['a'])
baskara['X2'] = ((baskara['b'] * -1) - (baskara['Delta'] ** 0.5)) / (2 * baskara['a'])

baskara['Xv'] = (baskara['b'] * -1) / (2 * baskara['a'])
baskara['Yv'] = (baskara['Delta'] * -1) / (4 * baskara['a'])

baskara['X Passo'] = (baskara['X2'] - baskara['X1']) / 100

baskara
```

```python
plt.close()

for values in np.array(baskara[['a', 'b', 'c', 'X1', 'X2', 'Xv', 'Yv', 'X Passo']]):
    # print(values)
    a, b, c, x1, x2, xv, yv, p = values
    
    x_list = []
    y_list = []
    
    x1 -= p
    
    for _ in range(101):
        x1 += p
        
        x_list.append(x1)
        
    for xb in x_list:
        y = (a * (xb) ** 2) + (b * xb) + c
        y_list.append(y)
    
    #print(x_list)
        
    plt.plot(x_list, y_list)
    plt.axvline(xv, color='r')
    plt.axhline(yv, color='g')
    plt.show()
```
